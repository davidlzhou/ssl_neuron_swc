```python
import numpy as np
import os

import json
import ssl_neuron
```

```python
configname = 'configs/config.json'
config = json.load(open(configname))
```

```python
SWC = [np.loadtxt(f'SWC_ALL/{file}', comments='#', delimiter=' ', dtype=str)[:, 1:].astype(float) for file in os.listdir('SWC_ALL')[:128]]
```

```python
train_loader, val_loader = ssl_neuron.datasets.build_dataloader(config, cells_or_SWC='SWC', SWC=SWC)
```

```python
model = ssl_neuron.graphdino.create_model(config)
trainer = ssl_neuron.train.Trainer(config, model, [train_loader, val_loader])
```

```python
print('Start training.')
trainer.train()
print('Done.')
```
